# Embedding Models Evaluation 

```python
import os
import pandas as pd
import datetime
from typing import List, Dict, Optional, Set
import torch
import gc

from sentence_transformers import SentenceTransformer

from langchain_ollama import OllamaEmbeddings
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores.faiss import FAISS
from langchain.embeddings.base import Embeddings

##################################################
# Custom Embeddings Class for SentenceTransformer
##################################################
class CustomSentenceTransformerEmbeddings(Embeddings):
    """
    Allows using a SentenceTransformer model within a LangChain-based FAISS store.
    """

    def __init__(self, model: SentenceTransformer):
        self.model = model

    def embed_query(self, text: str) -> List[float]:
        return self.model.encode(text).tolist()

    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        return self.model.encode(texts).tolist()

    def unload_model(self):
        """
        Remove the model from memory after processing to free up GPU resources.
        """
        if self.model:
            del self.model  # Delete the model instance
            self.model = None  # Ensure the reference is cleared
            torch.cuda.empty_cache()  # Clear the GPU cache
            gc.collect()  # Run garbage collection
        else:
            print("[DEBUG] Embedding model was already None or not set.")


################################
# Evaluation Helper Functions
################################
def evaluate_retrieval(retrieved_chunks, expected_chunks, comparison_length=100):
    """
    Exact match check: compare based on the first `comparison_length` characters of each chunk.
    Also prints a quick overview.
    """
    expected_prefixes = set([c[:comparison_length].strip() for c in expected_chunks])
    retrieved_prefixes = set([c[:comparison_length].strip() for c in retrieved_chunks])

    correctly_retrieved = expected_prefixes.intersection(retrieved_prefixes)
    missed_chunks = expected_prefixes.difference(retrieved_prefixes)

    print(f"\nExact-Match Evaluation (First {comparison_length} characters):")
    print(
        f"  Correctly retrieved chunks: {len(correctly_retrieved)}/{len(expected_chunks)}"
    )
    for prefix in correctly_retrieved:
        print(f"    ✔ Retrieved Prefix: {prefix}")

    print(f"  Missed chunks: {len(missed_chunks)}/{len(expected_chunks)}")
    for prefix in missed_chunks:
        print(f"    ✘ Missed Prefix: {prefix}")

    return correctly_retrieved, missed_chunks


def evaluate_ranked_retrieval(docs_with_ranks, expected_chunks, comparison_length=100):
    """
    Rank-based check: for each expected chunk, see at which rank its prefix appears
    in the retrieved list, then compute Mean Rank and MRR.
    Also prints a quick overview.
    """
    ranks = []
    for chunk in expected_chunks:
        chunk_prefix = chunk[:comparison_length].strip()
        matched_positions = [
            rank
            for (rank, doc) in docs_with_ranks
            if doc.page_content[:comparison_length].strip() == chunk_prefix
        ]
        if matched_positions:
            ranks.append(matched_positions[0])
        else:
            ranks.append(None)

    found_ranks = [r for r in ranks if r is not None]
    missed_count = sum(r is None for r in ranks)

    if found_ranks:
        mean_rank = sum(found_ranks) / len(found_ranks)
        mrr = sum((1.0 / r) for r in found_ranks) / len(found_ranks)
    else:
        mean_rank = None
        mrr = None

    print(f"\nRanked Evaluation (First {comparison_length} characters):")
    print(f"  Found: {len(found_ranks)}/{len(expected_chunks)}")
    print(f"  Missed: {missed_count}/{len(expected_chunks)}")
    if mean_rank is not None:
        print(f"  Mean Rank: {mean_rank:.2f}")
    else:
        print(f"  Mean Rank: N/A")
    if mrr is not None:
        print(f"  MRR: {mrr:.3f}")
    else:
        print(f"  MRR: N/A")

    return mean_rank, mrr


def save_csv(evaluation_data, k_value, output_csv_path):
    df = pd.DataFrame(evaluation_data)
    df["k_value"] = k_value

    # Columns in a consistent order
    columns_order = [
        "model",
        "attribute",
        "k_value",
        "retrieved_docs",
        "found_exact",
        "missed_exact",
        "mean_rank",
        "mrr",
    ]
    df = df.reindex(columns=columns_order, fill_value="N/A")

    # Export to CSV
    os.makedirs(os.path.dirname(output_csv_path), exist_ok=True)
    if not os.path.exists(output_csv_path):
        df.to_csv(output_csv_path, index=False, mode="w", encoding="utf-8")
    else:
        df.to_csv(
            output_csv_path, index=False, mode="a", header=False, encoding="utf-8"
        )


############################################################
# Main function: compare_models with per-attribute queries
############################################################
def compare_models(
    attributes: List[str],
    vectorstore_paths: Dict[str, str],
    schema: str,
    attribute_to_expected_chunk: Dict[str, str],
    k_value: int = 24,
    output_csv_path: str = "../data/rerank_eval/reranker_eval.csv",
    device: str = "cuda",
    comparison_length: int = 100,
    embedding_models: Optional[Dict[str, Embeddings]] = None,
):
    """
    Compare multiple models for a given schema by retrieving documents
    FOR EACH ATTRIBUTE individually, then evaluate with exact-match & rank-based metrics.
    Finally, print a Markdown table & export to CSV.
    """

    if embedding_models is None:
        embedding_models = {}

    evaluation_data = []

    for embedding_model_name, faiss_path in vectorstore_paths.items():
        print(f"\nRetrieving documents using {embedding_model_name}...")

        # Choose embeddings
        if embedding_model_name == "jinaai/jina-embeddings-v3":
            if embedding_model_name not in embedding_models:
                model = SentenceTransformer(
                    embedding_model_name,
                    trust_remote_code=True,
                    revision="main",
                    device=device,
                    model_kwargs={
                        "use_flash_attn": False,
                        # "attn_implementation": "flash_attention_2",
                        # "torch_dtype": torch.bfloat16,
                        #"default_task": "text-matching"
                        },
                )
                embeddings = CustomSentenceTransformerEmbeddings(model=model)
                embedding_models[embedding_model_name] = embeddings
                print(f"Loaded embedding model: {embedding_model_name}")
            else:
                embeddings = embedding_models[embedding_model_name]
                print(f"Reusing embedding model: {embedding_model_name}")
        elif embedding_model_name in [
            "HIT-TMG/KaLM-embedding-multilingual-mini-instruct-v1.5",
        ]:
            if embedding_model_name not in embedding_models:
                model = SentenceTransformer(
                    embedding_model_name,
                    trust_remote_code=False, # Do NOT set trust_remote_code
                    revision="main",
                    local_files_only=True,
                    device=device,
                    # prompts={
                    #     "Instruct": "Find the most similar attributes.",
                    #     "Query": "",     
                    #     }
                )
                embeddings = CustomSentenceTransformerEmbeddings(model=model)
                embedding_models[embedding_model_name] = embeddings
                print(f"Loaded embedding model: {embedding_model_name}")
            else:
                embeddings = embedding_models[embedding_model_name]
                print(f"Reusing embedding model: {embedding_model_name}")
        else:
            # embeddings = OllamaEmbeddings(model=embedding_model_name)
            if embedding_model_name not in embedding_models:
                embeddings = OllamaEmbeddings(model=embedding_model_name)
                embedding_models[embedding_model_name] = embeddings
                print(f"Loaded embedding model: {embedding_model_name}")
            else:
                embeddings = embedding_models[embedding_model_name]
                print(f"Reusing embedding model: {embedding_model_name}")

        # Load vector store
        vectorstore = FAISS.load_local(
            faiss_path,
            embeddings=embeddings,
            allow_dangerous_deserialization=True,
        )

        # For each attribute, do a separate query
        for attribute in attributes:
            query = attribute
            expected_chunk = attribute_to_expected_chunk.get(attribute)
            if not expected_chunk:
                print(
                    f"[Warning] No expected chunk found for attribute '{attribute}'. Skipping."
                )
                continue

            print(f"\n[Attribute: {attribute}] Query = '{query}'")

            # Example MMR-based retriever
            retriever = vectorstore.as_retriever(
                search_type="mmr",
                search_kwargs={
                    "filter": {"schema_type": schema},
                    "k": k_value,
                    "fetch_k": 100,
                    "lambda_mult": 1,
                },
            )

            retrieved_docs = retriever.invoke(query)
            print(f"Number of retrieved documents: {len(retrieved_docs)}")

            # Example Similarity Score-based retriever

            # Evaluate retrieval
            retrieved_chunks = [doc.page_content for doc in retrieved_docs]
            found_exact, missed_chunks = evaluate_retrieval(
                retrieved_chunks, [expected_chunk], comparison_length=comparison_length
            )

            # Evaluate ranking
            docs_with_ranks = list(enumerate(retrieved_docs, start=1))
            mean_rank, mrr = evaluate_ranked_retrieval(
                docs_with_ranks, [expected_chunk], comparison_length=comparison_length
            )

            # Print top docs
            top_docs = retrieved_chunks[:10]
            print(f"\n[Original Top Docs (first {len(top_docs)})]")
            for i, doc_text in enumerate(top_docs, start=1):
                print(f"  {i}. {doc_text[:80].replace('\n',' ')}...")

            # Collect results
            evaluation_data.append(
                {
                    "model": embedding_model_name,
                    "attribute": attribute,
                    "retrieved_docs": len(retrieved_docs),
                    "found_exact": len(found_exact),
                    "missed_exact": len(missed_chunks),
                    "mean_rank": mean_rank,
                    "mrr": mrr,
                }
            )

    # Save CSV
    save_csv(evaluation_data, k_value, output_csv_path)


#########
# Usage
#########
if __name__ == "__main__":
    k_values = [2, 5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
    # k_values = [2, 5, 10,]
    embeddings_name = "row_cs2000_co0_faiss_index_COS"
    embeddings_name_512 = "row_cs2000_co0_faiss_index_COS"
    vectorstore_paths = {
        "jinaai/jina-embeddings-v3": f"../embeddings/jinaai_jina-embeddings-v3/{embeddings_name}",
        "HIT-TMG/KaLM-embedding-multilingual-mini-instruct-v1.5": f"../embeddings/HIT-TMG_KaLM-embedding-multilingual-mini-instruct-v1.5/{embeddings_name}",
        # "bge-m3:latest": f"../embeddings/bge-m3/{embeddings_name}", # Ollama
        # "snowflake-arctic-embed2:latest": f"../embeddings/snowflake/{embeddings_name}", # Ollama
        # "jina/jina-embeddings-v2-base-de:latest": f"../embeddings/jina_jina/{embeddings_name}", # Ollama
        # "paraphrase-multilingual:latest": f"../embeddings/paraphrase/{embeddings_name}", # Ollama
        # "jeffh/intfloat-multilingual-e5-large-instruct:f32": f"../embeddings/jeffh_intfloat/{embeddings_name_512}", # Ollama
        # "granite-embedding:278m": f"../embeddings/granite/{embeddings_name_512}", # Ollama
        # "bge-large:latest": f"../embeddings/bge-large/{embeddings_name_512}", # Ollama
        # "mxbai-embed-large:latest": f"../embeddings/mxbai/{embeddings_name_512}", # Ollama
    }

    # Prepare CSV output path
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    output_csv_path = f"../data/embed_eval/embed_eval_res_{timestamp}_test.csv"

    # Initialize a dictionary to keep track of embedding models for unloading later
    embedding_models = {}

    # Run with multiple k_values
    for k_value in k_values:
        for schema, attributes in attribute_schema_mapping.items():
            compare_models(
                attributes=attributes,
                vectorstore_paths=vectorstore_paths,
                schema=schema,
                attribute_to_expected_chunk=attribute_to_expected_chunk,
                k_value=k_value,
                output_csv_path=output_csv_path,
                device="cuda",
                comparison_length=100,
                embedding_models=embedding_models,
            )
    
    
    # Unload Embedding Models to Free GPU Memory
    print("\n>>> Unloading Embedding Models from GPU Memory...")
    for embedding_model_name, embedding in embedding_models.items():
        if isinstance(embedding, CustomSentenceTransformerEmbeddings):
            embedding.unload_model()
            print(f"Unloaded embedding model: {embedding_model_name}")
        elif hasattr(embedding, "unload_model"):
            embedding.unload_model()
            print(f"Unloaded embedding model: {embedding_model_name}")
    print(">>> All embedding models have been unloaded.")

    print(f"\n>>> Completed. See results in {output_csv_path}.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def visualize_embedding_results(csv_file_path):
    """
    Loads the CSV results of your embedding model comparisons and
    creates multiple visualizations along with textual summaries.
    """
    # 1) Load Data
    df = pd.read_csv(csv_file_path)

    # A. Basic Cleanup & Info
    print(f"Loaded {len(df)} rows from {csv_file_path}")
    print("\n--- First 5 Rows of the DataFrame ---")
    print(df.head())

    # B. Convert columns to numeric (MRR, mean_rank, etc.) if needed
    numeric_cols = ["k_value", "retrieved_docs", "found_exact", "missed_exact", "mean_rank", "mrr"]
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    # C. Visual Setup
    sns.set(style="whitegrid", font_scale=1.1)

    ############################################################################
    # 2) Visualization Examples with Textual Results
    ############################################################################

    # (1) Bar Chart of found_exact vs. missed_exact across Models
    plt.figure(figsize=(12, 6))
    subset = df.groupby("model", as_index=False)[["found_exact", "missed_exact"]].sum()
    print("\n--- Found Exact vs. Missed Exact per Model ---")
    print(subset)
    # Plot side-by-side bar chart
    X = range(len(subset))
    width = 0.35
    plt.bar([x - width/2 for x in X], subset["found_exact"], width=width, label="found_exact")
    plt.bar([x + width/2 for x in X], subset["missed_exact"], width=width, label="missed_exact")
    plt.xticks(X, subset["model"], rotation=45, ha="right")
    plt.title("Found vs. Missed Chunks (Summed over all k_values)")
    plt.xlabel("Models")
    plt.ylabel("Count of Chunks")
    plt.legend()
    plt.tight_layout()
    plt.show()

    # (2) Mean Rank vs. k_value for each model (Line Plot)
    plt.figure(figsize=(12, 6))
    mean_rank_summary = {}
    for model_name, model_data in df.groupby("model"):
        meanranks = model_data.groupby("k_value")["mean_rank"].mean().reset_index()
        mean_rank_summary[model_name] = meanranks
        plt.plot(meanranks["k_value"], meanranks["mean_rank"], marker='o', label=model_name)
    plt.title("Mean Rank vs. k_value for each Model")
    plt.xlabel("k_value")
    plt.ylabel("Mean Rank")
    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

    print("\n--- Mean Rank per k_value for Each Model ---")
    for model, summary in mean_rank_summary.items():
        print(f"\nModel: {model}")
        print(summary.to_string(index=False))

    # (3) Scatter Plot: retrieved_docs vs. MRR
    plt.figure(figsize=(12, 6))
    df_scatter = df.dropna(subset=["mrr"])
    sns.scatterplot(
        data=df_scatter,
        x="retrieved_docs",
        y="mrr",
        hue="model",
        style="model",
        s=100
    )
    plt.title("Retrieved Docs vs. MRR Across Models")
    plt.xlabel("Retrieved Docs")
    plt.ylabel("MRR")
    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

    # Calculate and print correlation
    correlation = df_scatter[['retrieved_docs', 'mrr']].corr().iloc[0,1]
    print(f"\nCorrelation between Retrieved Docs and MRR: {correlation:.2f}")

    # (4) Heatmap: Correlation between numeric columns
    plt.figure(figsize=(8, 6))
    numeric_df = df[["found_exact", "missed_exact", "mean_rank", "mrr", "retrieved_docs"]].copy()
    corr_matrix = numeric_df.corr()
    sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
    plt.title("Correlation Heatmap of Retrieval Metrics")
    plt.tight_layout()
    plt.show()

    print("\n--- Correlation Matrix ---")
    print(corr_matrix)

    # (5) Missed Chunks per k_value for each Model
    plt.figure(figsize=(12, 6))
    missed_summary = {}
    for model_name, model_data in df.groupby("model"):
        missed_per_k = model_data.groupby("k_value")["missed_exact"].sum().reset_index()
        missed_summary[model_name] = missed_per_k
        plt.plot(missed_per_k["k_value"], missed_per_k["missed_exact"], marker='o', label=model_name)
    plt.title("Missed Chunks vs. k_value for each Model")
    plt.xlabel("k_value")
    plt.ylabel("Missed Chunks")
    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left")
    plt.tight_layout()
    plt.show()

    print("\n--- Missed Chunks per k_value for Each Model ---")
    for model, summary in missed_summary.items():
        print(f"\nModel: {model}")
        print(summary.to_string(index=False))


def interpret_results(csv_file_path):
    """
    Determines the minimum k_value at which each model
    has zero total missed chunks (summing across all attributes).
    """
    df = pd.read_csv(csv_file_path)

    # Convert to numeric
    numeric_cols = ["k_value", "found_exact", "missed_exact"]
    for col in numeric_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    # Sum missed_exact by (model, k_value)
    group_sums = df.groupby(["model", "k_value"])["missed_exact"].sum().reset_index()

    # For each model, find the smallest k_value where missed_exact == 0
    results = {}
    for model in group_sums["model"].unique():
        subset = group_sums[group_sums["model"] == model]
        # Find k_values where missed_exact is 0
        zero_miss_df = subset[subset["missed_exact"] == 0]
        if not zero_miss_df.empty:
            min_k = zero_miss_df["k_value"].min()
            results[model] = min_k
        else:
            results[model] = None

    print("\n====== MINIMUM k_value WITH ZERO MISSES (Summed) ======")
    for model, k_val in results.items():
        if k_val is None:
            print(f"{model}: No k_value found with zero total misses.")
        else:
            print(f"{model}: k_value = {k_val}")

    # Summary of Best Model
    # Filter models that have a valid k_value
    valid_models = {model: k for model, k in results.items() if k is not None}
    if valid_models:
        # Find the model with the smallest k_value
        best_model = min(valid_models, key=valid_models.get)
        best_k = valid_models[best_model]
        print("\n====== SUMMARY ======")
        print(f"The best model based on your criteria is '{best_model}' with k_value = {best_k}")
    else:
        print("\nNo model retrieved all documents correctly at any k_value.")


if __name__ == "__main__":
    csv_file = "../data/embed_eval/embed_eval_res_20250119_150332_test.csv"

    # embed_eval_res_20250114_154453
    #   - English
    #   - Jina no task
    # embed_eval_res_20250114_165318
    #   - all models
    #   - German and English
    #   - Jina no task
    # embed_eval_res_20250114_175749 
    #   - German and English
    #   - Jina with task=text-matching

    # 1) Visualize
    visualize_embedding_results(csv_file)

    # 2) Interpret
    interpret_results(csv_file)
```

# Reranker Models Evaluation

```python
import os
import datetime
import pandas as pd
from typing import List, Dict, Optional, Set
import torch
import gc

# Sentence-Transformers
from sentence_transformers import SentenceTransformer, CrossEncoder

# LangChain & Community modules
from langchain_ollama import OllamaEmbeddings
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
from langchain.embeddings.base import Embeddings


###############################################################################
# 1) Custom SentenceTransformer Embeddings Wrapper
###############################################################################
class CustomSentenceTransformerEmbeddings(Embeddings):
    """
    Allows using a SentenceTransformer model within a LangChain-based FAISS store.
    """

    def __init__(self, model: SentenceTransformer):
        self.model = model

    def embed_query(self, text: str) -> List[float]:
        return self.model.encode(text).tolist()

    def embed_documents(self, texts: List[str]) -> List[List[float]]:
        return self.model.encode(texts).tolist()

    def unload_model(self):
        """
        Remove the model from memory after processing to free up GPU resources.
        """
        if self.model:
            del self.model  # Delete the model instance
            self.model = None  # Ensure the reference is cleared
            torch.cuda.empty_cache()  # Clear the GPU cache
            gc.collect()  # Run garbage collection
        else:
            print("[DEBUG] Embedding model was already None or not set.")


###############################################################################
# 2) Base Reranker Class (CrossEncoder)
###############################################################################
class Reranker:
    """
    A reranker that uses a CrossEncoder to score (query, doc) pairs and reorder them.
    """

    def __init__(self, model_name: str, device: str = "cuda"):
        self.model_name = model_name
        self.reranker = CrossEncoder(
            model_name,
            automodel_args={"torch_dtype": "auto"},
            trust_remote_code=True,
            device=device,
            revision="main",            
        )

    def rerank(
        self, query: str, documents: List[str], top_n: Optional[int] = None
    ) -> List[Dict]:
        """
        Return a list of dicts: [{"text": doc_text, "score": float, "rank": int}, ...],
        sorted descending by score. If top_n is given, truncate the list.
        """
        if not documents:
            return []
        pairs = [[query, doc] for doc in documents]
        scores = self.reranker.predict(pairs)  # shape: [num_docs]

        # Sort by descending score
        doc_score_pairs = sorted(
            zip(documents, scores), key=lambda x: x[1], reverse=True
        )
        # Build result
        results = []
        for i, (text, score) in enumerate(doc_score_pairs):
            results.append({"text": text, "score": float(score), "rank": i + 1})
        if top_n is not None:
            results = results[:top_n]
        return results

    def evaluate_rerank(
        self,
        expected_chunk: str,
        reranked_results: List[Dict],
        comparison_length: int = 100,
    ) -> Dict:
        """
        Evaluate how well the reranker found the `expected_chunk`.
        Returns: {'found': bool, 'rank': int or None, 'mrr': float}
        """
        target_prefix = expected_chunk[:comparison_length].strip()
        found_rank = None
        for item in reranked_results:
            doc_prefix = item["text"][:comparison_length].strip()
            if doc_prefix == target_prefix:
                found_rank = item["rank"]
                break

        if found_rank is None:
            return {"found": False, "rank": None, "mrr": 0.0}
        else:
            return {"found": True, "rank": found_rank, "mrr": 1.0 / found_rank}

    def unload_model(self):
        """
        Remove the model from memory after processing to free up GPU resources.
        """
        if self.reranker:
            del self.reranker  # Delete the model instance
            self.reranker = None  # Ensure the reference is cleared
            torch.cuda.empty_cache()  # Clear the GPU cache
            gc.collect()  # Run garbage collection


###############################################################################
# 3) Utility: Evaluate Original Retrieval
###############################################################################
def evaluate_retrieval(
    retrieved_chunks: List[str],
    expected_chunks: List[str],
    comparison_length: int = 100,
):
    """
    Compare the first 'comparison_length' chars between retrieved vs. expected chunk(s)
    for an exact-match check.
    """
    expected_prefixes = set(c[:comparison_length].strip() for c in expected_chunks)
    retrieved_prefixes = set(c[:comparison_length].strip() for c in retrieved_chunks)

    correctly_retrieved = expected_prefixes.intersection(retrieved_prefixes)
    missed_chunks = expected_prefixes.difference(retrieved_prefixes)

    return correctly_retrieved, missed_chunks


def evaluate_ranked_retrieval(
    docs_with_ranks: List, expected_chunks: List[str], comparison_length: int = 100
):
    """
    For each expected chunk, see where it appears among docs_with_ranks (rank, doc).
    Then compute mean_rank, mrr, etc.
    """
    ranks = []
    for chunk in expected_chunks:
        chunk_prefix = chunk[:comparison_length].strip()
        matched_positions = [
            rank
            for (rank, doc) in docs_with_ranks
            if doc.page_content[:comparison_length].strip() == chunk_prefix
        ]
        ranks.append(matched_positions[0] if matched_positions else None)

    found_ranks = [r for r in ranks if r is not None]
    missed_count = sum(r is None for r in ranks)
    total_expected = len(expected_chunks)

    if found_ranks:
        mean_rank = sum(found_ranks) / len(found_ranks)
        mrr = sum(1.0 / r for r in found_ranks) / len(found_ranks)
    else:
        mean_rank = None
        mrr = None

    return {
        "found": len(found_ranks),
        "missed": missed_count,
        "mean_rank": mean_rank,
        "mrr": mrr,
    }


###############################################################################
# 4) Build Context from Reranked Results
###############################################################################
def build_context(reranked_results: List[Dict], threshold: float = 0.2) -> Set[str]:
    """
    From a reranked list, build a 'context' set by including docs whose
    score is within 'threshold' of top doc's score.
    Returns a set of doc strings (unique).
    """
    print("\nInclude chunk in context treshold:", threshold)

    if not reranked_results:
        return set()

    sorted_docs = sorted(reranked_results, key=lambda x: x["score"], reverse=True)
    top_score = sorted_docs[0]["score"]
    context_set = set()
    for item in sorted_docs:
        score_diff = top_score - item["score"]
        if score_diff <= threshold:
            context_set.add(item["text"])
        else:
            break
    return context_set


###############################################################################
# 5) Inspection Function to Verify Context Contains All Expected Chunks
###############################################################################
def inspect_context(
    context_map: Dict[str, Set[str]],
    attribute_to_expected_chunk: Dict[str, str],
    k_value: int,
    comparison_length: int = 100,
):
    """
    Inspect if the first 'comparison_length' characters of every expected chunk
    are present within the context for each model.
    """
    print(f"\n## Inspection of Contexts for Expected Chunks (First {comparison_length} Characters)\n")   

    for reranker_name, context_set in context_map.items():
        print(f"\n### Model: {reranker_name} k_value: {k_value}")
        all_found = True

        # Track missed chunks per reranker model
        missing_chunks = set()

        for attribute, expected_chunk_str in attribute_to_expected_chunk.items():
            expected_prefix = expected_chunk_str[:comparison_length].strip()

            found = any(
                expected_prefix in snippet[:comparison_length]
                for snippet in context_set
            )
            status = "FOUND" if found else "MISSING"

            if status == "MISSING":
                missing_chunks.add(expected_prefix)

            if not found:
                all_found = False

            print(f"  - Attribute: {attribute} => {status}")
            print(f'    • Expected Prefix: "{expected_prefix}"')

        if all_found:
            print("  **All expected chunks are present in the context.**\n")
        else:
            print("  **Some expected chunks are missing in the context.**")
            for missing_chunk in missing_chunks:
                print(f"     - {missing_chunk}")


###############################################################################
# 6) Print Debug Output for Original Retrieval + Reranking Steps
###############################################################################
def print_retrieval_debug(
    attribute: str,
    k_value: int,
    retrieved_chunks: List[str],
    retrieval_eval: Dict,
    comparison_length: int,
    correctly_retrieved: Set[str],
    missed_chunks: Set[str],
):
    # Print original retrieval debug info (like old code)
    print(f"\n=== Attribute: {attribute} | k={k_value} ===")
    print(f"Query: {attribute}")
    print(f"Retrieved docs: {len(retrieved_chunks)}")

    # Exact-match
    found_count = retrieval_eval["found"]
    missed_count = retrieval_eval["missed"]
    total = found_count + missed_count

    print(f"\nExact-Match Evaluation (First {comparison_length} chars):")
    print(f"  Correctly retrieved: {found_count}/{total}")
    for prefix in correctly_retrieved:
        print(f"    ✔ {prefix}")
    print(f"  Missed: {missed_count}/{total}")
    for prefix in missed_chunks:
        print(f"    ✘ {prefix}")


def print_ranking_debug(rank_eval: Dict):
    # Print old style ranking debug
    print("\nRanked Evaluation:")
    found = rank_eval["found"]
    missed = rank_eval["missed"]
    total = found + missed

    mean_rank = rank_eval["mean_rank"]
    mrr = rank_eval["mrr"]

    print(f"  Found: {found}/{total}")
    print(f"  Missed: {missed}/{total}")
    print(f"  Mean Rank: {mean_rank if mean_rank else 'N/A'}")
    print(f"  MRR: {mrr if mrr else 'N/A'}")


###############################################################################
# 7) Print Reranker Debug
###############################################################################
def print_reranking_debug(
    reranker_label: str,
    reranked: List[Dict],
    context_set: Set[str],
    k_value: int,
):
    # Show top 5 docs from reranked
    print(f"\n[Using Reranker: {reranker_label}]")
    for item in reranked[:5]:  # Change to k_value instead of 5
    # for item in reranked[:k_value]:  # Change to k_value instead of 5
        snippet_preview = item["text"][:70].replace("\n", " ")
        print(f"  R{item['rank']} | Score={item['score']:.4f} | {snippet_preview}...")

    # Show the context set
    print("\n[Context Set]:")
    for chunk in context_set:
        snippet_preview = chunk[:70].replace("\n", " ")
        print(f"  - {snippet_preview}...")


###############################################################################
# 8) Master function to compare reranking
###############################################################################
def evaluate_rerankers_for_attributes(
    attributes: List[str],
    vectorstore_paths: Dict[str, str],
    schema: str,
    attribute_to_expected_chunk: Dict[str, str],
    context_map: Dict[str, Set[str]],
    reranker_model_names: List[str],
    context_threshold: float,
    k_value: int = 24,
    output_csv_path: str = "../data/rerank_eval/reranker_eval.csv",
    device: str = "cuda",
    comparison_length: int = 100,
    embedding_models: Optional[Dict[str, Embeddings]] = None,
):
    """
    For each embedding 'model_name' in vectorstore_paths:
      1) Load the vector store for retrieval (via MMR).
      2) For each attribute, do retrieval -> measure old retrieval performance
         (exact match, mean rank, MRR).
      3) For each reranker in 'reranker_model_names', do rerank, measure new
         metrics (context size, rerank rank, etc.).
      4) Save everything in one CSV.
    """

    if embedding_models is None:
        embedding_models = {}

    all_rows = []

    for embedding_model_name, faiss_path in vectorstore_paths.items():
        print(f"\n[Embedding Model: {embedding_model_name}] => {faiss_path}")

        # Initialize or retrieve the embedding model
        if embedding_model_name == "jinaai/jina-embeddings-v3":
            if embedding_model_name not in embedding_models:
                st_model = SentenceTransformer(
                    embedding_model_name,
                    trust_remote_code=True,
                    device=device,
                    model_kwargs={"use_flash_attn": False},
                )
                embeddings = CustomSentenceTransformerEmbeddings(st_model)
                embedding_models[embedding_model_name] = embeddings
                print(f"Loaded embedding model: {embedding_model_name}")
            else:
                embeddings = embedding_models[embedding_model_name]
                print(f"Reusing embedding model: {embedding_model_name}")
        elif embedding_model_name in [
            "HIT-TMG/KaLM-embedding-multilingual-mini-instruct-v1.5"
        ]:
            if embedding_model_name not in embedding_models:
                embeddings = HuggingFaceEmbeddings(
                    model_name=embedding_model_name,
                    model_kwargs={
                        "trust_remote_code": True,
                        "device": device,
                    },
                )
                embedding_models[embedding_model_name] = embeddings
                print(f"Loaded embedding model: {embedding_model_name}")
            else:
                embeddings = embedding_models[embedding_model_name]
                print(f"Reusing embedding model: {embedding_model_name}")
        else:
            if embedding_model_name not in embedding_models:
                embeddings = OllamaEmbeddings(model=embedding_model_name)
                embedding_models[embedding_model_name] = embeddings
                print(f"Loaded embedding model: {embedding_model_name}")
            else:
                embeddings = embedding_models[embedding_model_name]
                print(f"Reusing embedding model: {embedding_model_name}")

        # Load vectorstore
        vectorstore = FAISS.load_local(
            faiss_path, embeddings=embeddings, allow_dangerous_deserialization=True
        )

        # For each attribute, do MMR retrieval
        for attribute in attributes:
            expected_chunk = attribute_to_expected_chunk.get(attribute)
            if not expected_chunk:
                print(f"[Warning] No expected chunk for '{attribute}'. Skipping.")
                continue

            # Retrieve
            retriever = vectorstore.as_retriever(
                search_type="mmr",
                search_kwargs={
                    "filter": {"schema_type": schema},
                    "k": k_value,
                    "fetch_k": 100,
                    "lambda_mult": 1.0,
                },
            )
            docs = retriever.invoke(attribute)
            retrieved_chunks = [d.page_content for d in docs]

            # Evaluate old retrieval
            correctly_retrieved, missed_chunks = evaluate_retrieval(
                retrieved_chunks, [expected_chunk], comparison_length
            )
            docs_with_ranks = list(enumerate(docs, start=1))
            rank_eval = evaluate_ranked_retrieval(
                docs_with_ranks, [expected_chunk], comparison_length
            )

            # Print debug if you like
            retrieval_eval = {
                "found": len(correctly_retrieved),
                "missed": len(missed_chunks),
            }
            print_retrieval_debug(
                attribute,
                k_value,
                retrieved_chunks,
                retrieval_eval,
                comparison_length,
                correctly_retrieved,
                missed_chunks,
            )
            print_ranking_debug(rank_eval)

            # For each reranker_
            top_docs_for_rerank = retrieved_chunks[:10]
            # top_docs_for_rerank = retrieved_chunks[:k_value]
            # Print original top docs
            print(f"\n[Original Top Docs (first {len(top_docs_for_rerank)})]")
            for i, doc_text in enumerate(top_docs_for_rerank, start=1):
                print(f"  {i}. {doc_text[:80].replace('\n',' ')}...")
            for reranker_name in reranker_model_names:
                reranker = Reranker(reranker_name, device=device)
                print("\nLoad reranker model to GPU memory:", reranker.model_name)
                reranked = reranker.rerank(attribute, top_docs_for_rerank, top_n=None)

                # Create reranker model context entry
                if reranker_name not in context_map:
                    context_map[reranker_name] = set()

                # Evaluate rerank
                rerank_eval = reranker.evaluate_rerank(
                    expected_chunk, reranked, comparison_length
                )
                # Build context
                context_set = build_context(reranked, threshold=context_threshold)
                for snippet in context_set:
                    context_map[reranker_name].add(snippet)
                context_size = len(context_set)

                # Determine if expected chunk is in context by comparing prefixes
                expected_prefix = expected_chunk[:comparison_length].strip()
                context_contains_expected = any(
                    chunk[:comparison_length].strip() == expected_prefix
                    for chunk in context_set
                )

                # Print debug about reranking
                print_reranking_debug(reranker_name, reranked, context_set, k_value)

                # Gather row
                row = {
                    "embedding_model": embedding_model_name,
                    "attribute": attribute,
                    "k_value": k_value,
                    "retrieved_docs": len(docs),
                    "found_exact": retrieval_eval["found"],
                    "missed_exact": retrieval_eval["missed"],
                    "mean_rank": rank_eval["mean_rank"],
                    "mrr": rank_eval["mrr"],
                    "reranker_model": reranker_name,
                    "rerank_found": "Yes" if rerank_eval["found"] else "No",
                    "rerank_rank": (
                        rerank_eval["rank"] if rerank_eval["found"] else None
                    ),
                    "rerank_mrr": rerank_eval["mrr"] if rerank_eval["found"] else 0.0,
                    "context_size": context_size,
                    "context_contains_expected": context_contains_expected,
                }
                all_rows.append(row)

                # Unload the reranker model to free GPU memory
                print("\n")
                # print(f"[DEBUG] Now calling unload_model() for {reranker} ...")
                reranker.unload_model()
                print("Unload reranker model from GPU memory:", reranker.model_name)
                # print(f"[DEBUG] Done with unload_model() for {reranker}")

    # Save CSV
    os.makedirs(os.path.dirname(output_csv_path), exist_ok=True)
    df = pd.DataFrame(all_rows)

    new_file = not os.path.exists(output_csv_path)
    df.to_csv(
        output_csv_path,
        index=False,
        mode="a" if not new_file else "w",
        header=new_file,
        encoding="utf-8",
    )

    # After all k_values, inspect the context
    inspect_context(context_map, attribute_to_expected_chunk, k_value, comparison_length=100)

    # Print final context sets if desired
    print("\n## Context Sets per Model\n")
    for model_name, ctx_set in context_map.items():
        print(f"Model: {model_name} => {len(ctx_set)} chunks in context.")
        # for snippet in ctx_set:
        # print(f"  - {snippet}")

    print(f"\n>>> Done. All results appended to {output_csv_path}")


###############################################################################
# Example usage
###############################################################################
if __name__ == "__main__":
    # Example usage
    vectorstore_paths = {
        "jinaai/jina-embeddings-v3": "../embeddings/jinaai_jina-embeddings-v3/row_cs2000_co0_faiss_index_COS",
        # Add others if needed
        # "snowflake-arctic-embed2:latest": "../embeddings/snowflake/row_cs2000_co0_faiss_index_COS"
    }

    attribute_schema_mapping = {
        "EPD_DataSet": [
            "UUID",
            "Version",
            "Name (de)",
            "Name (en)",
        ],
    }

    attribute_to_expected_chunk = {
        "UUID": "'UUID des Datensatzes','UUID of Process data set','UUID','UUID','UUID des Datensatzes. Zusammen mit der Versionsnummer in \"Datensatzversion\" wird der Datensatz damit eindeutig identifiziert','~','Automatically generated Universally Unique Identifier of this data set. Together with the \"Data set version\", the UUID uniquely identifies each data set.'",
        "Version": "'Datensatzversion','Data set version','dataSetVersion','Version','Versionsnummer des Datensatzes','~','Version number of data set. First two digits refer to major updates, the second two digits to minor revisions and error corrections etc. The third three digits are intended for automatic and internal counting of versions during data set development. Together with the data set's UUID, the \"Data set version\" uniquely identifies each data set.'",
        "Name (de)": "'Name','Name','baseName','StringMultiLang','Allgemeiner Name des Produkts oder Systems','~','General descriptive name of the process and/or its main good(s) or service(s) and/or its level of processing.'",
        "Name (en)": "'Name','Name','baseName','StringMultiLang','Allgemeiner Name des Produkts oder Systems','~','General descriptive name of the process and/or its main good(s) or service(s) and/or its level of processing.'",
    }

    # Rerankers
    reranker_model_names = [
        "jinaai/jina-reranker-v2-base-multilingual",
        "BAAI/bge-reranker-v2-m3",
        "Alibaba-NLP/gte-multilingual-reranker-base",
        # etc.
    ]

    # Run multiple k_values:
    k_values = [10, 15, 20]

    # Store contexts in a dictionary
    context_map: Dict[str, Set[str]] = {}

    # CSV output
    timestamp = datetime.datetime.now().strftime("%Y%m%d_%H%M%S")
    output_csv_path = f"../data/rerank_eval/reranker_eval_res_{timestamp}_test.csv"

    # Initialize a dictionary to keep track of embedding models for unloading later
    embedding_models = {}

    # Now run everything
    for k_val in k_values:
        for schema, attributes in attribute_schema_mapping.items():
            evaluate_rerankers_for_attributes(
                attributes=attributes,
                vectorstore_paths=vectorstore_paths,
                schema=schema,
                attribute_to_expected_chunk=attribute_to_expected_chunk,
                context_map=context_map,
                reranker_model_names=reranker_model_names,
                context_threshold=0.1,
                k_value=k_val,
                output_csv_path=output_csv_path,
                device="cuda",
                comparison_length=100,
                embedding_models=embedding_models,
            )

    print(f"\n>>> Completed. See results in {output_csv_path}.")

    # Unload Embedding Models to Free GPU Memory
    print("\n>>> Unloading Embedding Models from GPU Memory...")
    for embedding_model_name, embedding in embedding_models.items():
        if isinstance(embedding, CustomSentenceTransformerEmbeddings):
            embedding.unload_model()
            print(f"Unloaded embedding model: {embedding_model_name}")
        elif hasattr(embedding, "unload_model"):
            embedding.unload_model()
            print(f"Unloaded embedding model: {embedding_model_name}")
    print(">>> All embedding models have been unloaded.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from scipy import stats

# 1. Load the CSV data
csv_path = "../data/rerank_eval/reranker_eval_res_20250117_174252.csv"  # Update with your CSV path
if not os.path.exists(csv_path):
    raise FileNotFoundError(f"The CSV file was not found at path: {csv_path}")

df = pd.read_csv(csv_path)

# 2. Data Cleaning and Preparation
df["context_contains_expected"] = df["context_contains_expected"].astype(bool)

# Replace NaN with appropriate values
df["rerank_rank"] = df["rerank_rank"].fillna(-1).astype(float)
df["rerank_mrr"] = df["rerank_mrr"].fillna(0.0)
df["mean_rank"] = df["mean_rank"].fillna(-1)
df["mrr"] = df["mrr"].fillna(0.0)
df["context_size"] = df["context_size"].fillna(0).astype(float)

# 3. Set Plot Aesthetics
sns.set_theme(style="whitegrid")
plt.rcParams.update({"figure.max_open_warning": 0})


# -------------------------------------------------------------------
# 4. Plot Mean Rank and MRR for Each Reranker
# -------------------------------------------------------------------
plt.figure(figsize=(14, 6))
metrics = ["mean_rank", "mrr"]
for i, metric in enumerate(metrics, 1):
    plt.subplot(1, 2, i)
    sns.barplot(
        data=df,
        x="reranker_model",
        y=metric,
        hue="reranker_model",
        palette="viridis",
        dodge=False,
    )
    plt.title(f'Average {metric.replace("_", " ").title()} by Reranker')
    plt.xlabel("Reranker Model")
    plt.ylabel(metric.replace("_", " ").title())
    plt.xticks(rotation=45)
    # Remove legend to avoid duplication
    plt.legend([], [], frameon=False)
plt.tight_layout()
plt.show()

# Textual Summary for Mean Rank and MRR
print("\n=== Mean Rank and MRR by Reranker Model ===")
grp_mean_rank = df.groupby("reranker_model")["mean_rank"].mean().reset_index()
grp_mrr = df.groupby("reranker_model")["mrr"].mean().reset_index()
summary_metrics = pd.merge(grp_mean_rank, grp_mrr, on="reranker_model")
summary_metrics = summary_metrics.rename(
    columns={"mean_rank": "Average Mean Rank", "mrr": "Average MRR"}
)
print(summary_metrics.to_string(index=False))

print("\nExplanation:")
print("   - This bar plot shows the average mean rank and MRR for each reranker model.")
print("   - Lower mean rank and higher MRR indicate better performance.")
print(
    "   - From the summary table, Model X has the lowest mean rank and highest MRR, indicating superior performance."
)


# -------------------------------------------------------------------
# 5. Plot: Context Size Distribution per Reranker
# -------------------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=df, x="reranker_model", y="context_size", hue="reranker_model", palette="Set2"
)
plt.title("Distribution of Context Sizes by Reranker")
plt.xlabel("Reranker Model")
plt.ylabel("Context Size (Number of Chunks)")
plt.xticks(rotation=45)
# Remove legend to avoid duplication
plt.legend([], [], frameon=False)
plt.show()

# Textual Summary for Context Size Distribution
print("\n=== Context Size Distribution by Reranker Model ===")
context_stats = df.groupby("reranker_model")["context_size"].describe().reset_index()
print(context_stats.to_string(index=False))

print("\nExplanation:")
print(
    "   - The boxplots illustrate the distribution of context sizes across different reranker models."
)
print(
    "   - Models with consistently lower context sizes are more efficient in utilizing fewer chunks."
)
print(
    "   - Statistical summaries show that Model Y has a lower median context size compared to others."
)


# -------------------------------------------------------------------
# 6. Plot: Success Rate of Finding All Expected Chunks
# -------------------------------------------------------------------
success_rate = (
    df.groupby("reranker_model")["context_contains_expected"].mean().reset_index()
)

plt.figure(figsize=(8, 6))
sns.barplot(
    data=success_rate,
    x="reranker_model",
    y="context_contains_expected",
    hue="reranker_model",
    palette="magma",
)
plt.title("Success Rate of Finding All Expected Chunks by Reranker")
plt.xlabel("Reranker Model")
plt.ylabel("Success Rate")
plt.ylim(0, 1)
plt.xticks(rotation=45)
# Remove legend to avoid duplication
plt.legend([], [], frameon=False)
plt.show()

# Textual Summary for Success Rate
print("\n=== Success Rate of Finding All Expected Chunks by Reranker ===")
success_rate["Success Rate (%)"] = success_rate["context_contains_expected"] * 100
success_rate = success_rate.drop(columns=["context_contains_expected"])
print(success_rate.to_string(index=False))


print("\nExplanation:")
print(
    "   - This bar plot displays the success rate of each reranker model in finding all expected chunks."
)
print(
    "   - Higher success rates are preferable as they indicate better retrieval accuracy."
)
print(
    "   - According to the summary, Model Z achieves a 100% success rate, outperforming other models."
)


# -------------------------------------------------------------------
# 7. Build a Summary DataFrame Including Performance at Different k Values
# -------------------------------------------------------------------
# Group by 'reranker_model' and 'k_value' to compute average context_size and success rate
summary_corr = (
    df.groupby(["reranker_model", "k_value"])
    .agg({"context_size": "mean", "context_contains_expected": "mean"})
    .reset_index()
)

# Rename columns for clarity
summary_corr = summary_corr.rename(
    columns={
        "context_size": "Average Context Size",
        "context_contains_expected": "Success Rate (%)",
    }
)

# Calculate 'Success Rate (%)'
summary_corr["Success Rate (%)"] = summary_corr["Success Rate (%)"] * 100

# Debug: Check the summary_corr DataFrame
# print("\n[DEBUG] summary_corr DataFrame after groupby and aggregation:\n", summary_corr.head())

# Check for required columns
required_columns = [
    "Average Context Size",
    "Success Rate (%)",
    "reranker_model",
    "k_value",
]
missing_columns = [col for col in required_columns if col not in summary_corr.columns]
if missing_columns:
    raise ValueError(f"Missing columns in 'summary_corr' DataFrame: {missing_columns}")

# -------------------------------------------------------------------
# 8. Scatter Plot: Correlation Between Context Size and Success Rate
# -------------------------------------------------------------------
plt.figure(figsize=(10, 8))
sns.scatterplot(
    data=summary_corr,
    x="Average Context Size",
    y="Success Rate (%)",
    hue="reranker_model",
    style="k_value",
    s=250,
    palette="deep",
    markers=["o", "s", "D", "^", "v", "P", "*", "X", "<", ">"],  # Adjust markers for different k_values
)

# Add regression lines with correct linestyle using line_kws
sns.regplot(
    data=summary_corr,
    x="Average Context Size",
    y="Success Rate (%)",
    scatter=False,
    fit_reg=True, # Disable to declutter
    color="gray", # Visible if fit_reg=True
    line_kws={"linestyle": "--", "alpha": 0.2}, # Visible if fit_reg=True
)

# Annotate points with model names and k_values
# for idx, row in summary_corr.iterrows():
#     plt.text(
#         row["Average Context Size"] + 0.05,
#         row["Success Rate (%)"] - 2,
#         f"{row['reranker_model']} (k={int(row['k_value'])})",
#         horizontalalignment="left",
#         size="medium",
#         color="black",
#         weight="semibold",
#     )

# Set axis limits dynamically based on data
plt.title("Correlation Between Context Size and Success Rate")
plt.xlabel("Average Context Size (Number of Chunks)")
plt.ylabel("Success Rate (%)")
plt.ylim(
    max(0, summary_corr["Success Rate (%)"].min() - 5),
    min(101, summary_corr["Success Rate (%)"].max() + 5),
)
plt.xlim(
    max(0, summary_corr["Average Context Size"].min() - 1),
    summary_corr["Average Context Size"].max() + 1,
)
plt.legend(title="Reranker Model & k", loc="lower right", ncol=2, markerscale=0.8)
plt.grid(True)
plt.tight_layout()
plt.show()

# Calculate and print Pearson Correlation for context_size vs Success Rate
corr_cs_sr = (
    summary_corr[["Average Context Size", "Success Rate (%)"]].corr().iloc[0, 1]
)
print(f"\nPearson Correlation between Context Size and Success Rate: {corr_cs_sr:.2f}")


print("\nExplanation:")
print(
    "   - The scatter plot with regression lines shows the relationship between average context size and success rate."
)
print(
    "   - A positive correlation suggests that larger context sizes may contribute to higher success rates."
)
print(
    "   - The Pearson correlation coefficient of {:.2f} indicates a strong {} correlation.".format(
        corr_cs_sr, "positive" if corr_cs_sr > 0 else "negative"
    )
)

# -------------------------------------------------------------------
# 9. Scatter Plot: MRR vs Mean Rank
# -------------------------------------------------------------------
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df,
    x="mean_rank",
    y="mrr",
    hue="reranker_model",
    style="reranker_model",
    s=150,
    palette="deep",
)
plt.title("MRR vs Mean Rank by Reranker")
plt.xlabel("Mean Rank")
plt.ylabel("MRR")
plt.legend(title="Reranker Model", loc="upper left")
plt.show()

# Textual Summary for MRR vs Mean Rank
print("\n=== Correlation Between MRR and Mean Rank ===")
corr_mr_mrr = df[["mean_rank", "mrr"]].corr().iloc[0, 1]
print(f"Overall Pearson Correlation between Mean Rank and MRR: {corr_mr_mrr:.2f}")

print("\nPearson Correlation by Reranker Model:")
for model in df["reranker_model"].unique():
    subset = df[df["reranker_model"] == model]
    if len(subset) >= 2:
        c = subset[["mean_rank", "mrr"]].corr().iloc[0, 1]
        print(f"  - {model}: {c:.2f}")
    else:
        print(f"  - {model}: Not enough data to compute correlation")

print("\nExplanation:")
print(
    "   - This scatter plot depicts the relationship between MRR and mean rank for each reranker model."
)
print("   - A higher MRR and lower mean rank signify better performance.")
print(
    "   - The overall correlation is {:.2f}, indicating a {} relationship.".format(
        corr_mr_mrr, "positive" if corr_mr_mrr > 0 else "negative"
    )
)

# -------------------------------------------------------------------
# 10. Heatmap of Average Context Size per Reranker and Attribute
# -------------------------------------------------------------------
heat_data = df.groupby(["reranker_model", "attribute"])["context_size"].mean().unstack()

plt.figure(figsize=(14, 10))
sns.heatmap(heat_data, annot=True, fmt=".1f", cmap="YlGnBu", linewidths=0.5)
plt.title("Average Context Size by Reranker and Attribute")
plt.xlabel("Attribute")
plt.ylabel("Reranker Model")
plt.xticks(rotation=45, ha="right")
plt.show()

# Textual Summary for Heatmap
print("\n=== Average Context Size by Reranker and Attribute ===")
print(heat_data.round(1).to_string())

print("\nExplanation:")
print(
    "   - The heatmap visualizes the average context size for each reranker model across different attributes."
)
print("   - Darker colors represent larger context sizes.")
print("   - It highlights which models require more chunks for specific attributes.")

# -------------------------------------------------------------------
# 11. Pairplot for Comprehensive Relationships
# -------------------------------------------------------------------
sns.pairplot(
    df,
    vars=["mean_rank", "mrr", "context_size", "rerank_mrr"],
    hue="reranker_model",
    palette="bright",
    markers=["o", "s", "D"],  # Adjust markers to match number of reranker models
)
plt.suptitle("Pairplot of Reranker Metrics", y=1.02)
plt.show()

print("\n=== Pairwise Relationships Among Metrics ===")
corr_matrix = df[["mean_rank", "mrr", "context_size", "rerank_mrr"]].corr()
print("Pearson Correlation Matrix:")
print(corr_matrix.to_string())

print("\n=== Insights from Pairwise Correlations ===")
for i, row in corr_matrix.iterrows():
    for j, val in row.items():
        if i != j and abs(val) > 0.5:
            relation = "positively" if val > 0 else "negatively"
            print(f"  - {i} is {relation} correlated with {j} (Correlation: {val:.2f})")

print("\nExplanation:")
print(
    "   - The pairplot provides a comprehensive view of the relationships between mean rank, MRR, context size, and rerank MRR."
)
print(
    "   - Significant correlations are identified, helping to understand how these metrics interact."
)

# -------------------------------------------------------------------
# 12. Performance at Different k Values
# -------------------------------------------------------------------
# Check if 'k_value' exists
if "k_value" not in df.columns:
    raise ValueError(
        "The DataFrame does not contain a 'k_value' column required for plotting performance at different k values."
    )

# Define the performance metric
performance_metric = (
    "Success Rate (%)"  # You can change this to another metric if desired
)

# Group by 'reranker_model' and 'k_value' to compute average performance
performance_summary = (
    df.groupby(["reranker_model", "k_value"])
    .agg({"context_contains_expected": "mean"})
    .reset_index()
)

# Calculate 'Success Rate (%)'
performance_summary["Success Rate (%)"] = (
    performance_summary["context_contains_expected"] * 100
)

# Drop the intermediate column
performance_summary = performance_summary.drop(columns=["context_contains_expected"])

# Debug: Check the performance_summary DataFrame
# print("\n[DEBUG] Performance Summary at Different k Values:\n", performance_summary.head())

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=performance_summary,
    x="k_value",
    y="Success Rate (%)",
    hue="reranker_model",
    marker="o",
)
plt.title("Average Success Rate at Different k Values by Reranker Model")
plt.xlabel("k")
plt.ylabel("Success Rate (%)")
plt.xticks(sorted(df["k_value"].unique()))
plt.legend(title="Reranker Model", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()

# Textual Summary for Performance at Different k Values
print("\n=== Performance Summary at Different k Values ===")
print(performance_summary.to_string(index=False))

print("\nExplanation:")
print(
    "   - The line plot shows how the success rate varies with different k values for each reranker model."
)
print("   - This helps in assessing the models' performance as the cutoff k increases.")
print(
    "   - Generally, as k increases, success rates may plateau or vary based on the model's efficiency."
)

# -------------------------------------------------------------------
# 13. Compare Context Size to k_value
# -------------------------------------------------------------------
# Check if 'k_value' exists
if "k_value" not in df.columns:
    raise ValueError(
        "The DataFrame does not contain a 'k_value' column required for plotting performance at different k values."
    )

# Group by 'reranker_model' and 'k_value' to compute average context_size and success rate
context_k_summary = (
    df.groupby(["reranker_model", "k_value"])
    .agg({"context_size": "mean", "context_contains_expected": "mean"})
    .reset_index()
)

# Rename columns for clarity
context_k_summary = context_k_summary.rename(
    columns={
        "context_size": "Average Context Size",
        "context_contains_expected": "Success Rate (%)",
    }
)

# Calculate 'Success Rate (%)'
context_k_summary["Success Rate (%)"] = context_k_summary["Success Rate (%)"] * 100

# Drop the intermediate 'context_contains_expected' if present
context_k_summary = context_k_summary.drop(
    columns=["context_contains_expected"], errors="ignore"
)

# Debug: Check the context_k_summary DataFrame
# print("\n[DEBUG] context_k_summary DataFrame after groupby and aggregation:\n", context_k_summary.head())

# Check for required columns
required_columns = [
    "Average Context Size",
    "Success Rate (%)",
    "reranker_model",
    "k_value",
]
missing_columns = [
    col for col in required_columns if col not in context_k_summary.columns
]
if missing_columns:
    raise ValueError(
        f"Missing columns in 'context_k_summary' DataFrame: {missing_columns}"
    )

# Plotting
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=context_k_summary,
    x="k_value",
    y="Average Context Size",
    hue="reranker_model",
    marker="o",
)
plt.title("Average Context Size at Different k Values by Reranker Model")
plt.xlabel("k Value")
plt.ylabel("Average Context Size (Number of Chunks)")
plt.xticks(sorted(df["k_value"].unique()))
plt.legend(title="Reranker Model", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.grid(True)
plt.tight_layout()
plt.show()

# Textual Summary for Context Size at Different k Values
print("\n=== Average Context Size at Different k Values by Reranker Model ===")
print(context_k_summary.to_string(index=False))


print("\nExplanation:")
print(
    "   - This line plot illustrates how the average context size changes with different k values for each reranker model."
)
print(
    "   - It provides insights into how context size scales with k, which can impact computational resources."
)

# -------------------------------------------------------------------
# 14. Summary Statistics Table & Best Model
# -------------------------------------------------------------------
# Build the final summary DataFrame with all necessary metrics
final_summary = (
    df.groupby("reranker_model")
    .agg(
        {
            "mean_rank": "mean",
            "mrr": "mean",
            "context_size": "mean",
            "rerank_mrr": "mean",
            "context_contains_expected": "mean",
        }
    )
    .reset_index()
)

print("\n=== Summary Statistics by Reranker Model ===")
print(final_summary.to_string(index=False))

# Define the criteria: highest success rate and lowest context size
best_summary = final_summary.copy()
best_summary["Success Rate (%)"] = best_summary["context_contains_expected"] * 100
best_summary = best_summary.rename(
    columns={
        "mean_rank": "Average Mean Rank",
        "mrr": "Average MRR",
        "context_size": "Average Context Size",
        "rerank_mrr": "Average Rerank MRR",
    }
)

# Sort by Success Rate descending and Average Context Size ascending
best_summary_sorted = best_summary.sort_values(
    by=["Success Rate (%)", "Average Context Size"], ascending=[False, True]
).reset_index(drop=True)

# Identify the top model(s) with the highest success rate
top_success_rate = best_summary_sorted["Success Rate (%)"].max()
top_models = best_summary_sorted[
    best_summary_sorted["Success Rate (%)"] == top_success_rate
]

# Among top models, find the one(s) with the smallest context size
min_context_size = top_models["Average Context Size"].min()
best_models = top_models[top_models["Average Context Size"] == min_context_size]

print("\n=== Best Reranker Model(s) ===")
for idx, row in best_models.iterrows():
    print(f"\nModel: {row['reranker_model']}")
    print(f"  Success Rate: {row['Success Rate (%)']:.2f}%")
    print(f"  Average Context Size: {row['Average Context Size']:.1f} chunks")
    print(f"  Average MRR: {row['Average MRR']:.3f}")
    print(f"  Average Rerank MRR: {row['Average Rerank MRR']:.3f}")
    print(f"  Average Mean Rank: {row['Average Mean Rank']:.1f}")

if len(best_models) > 1:
    print(
        "\nNote: Multiple models share the highest success rate and lowest context size."
    )
else:
    print("\nNote: Only one model meets the criteria as the best-performing model.")

print("\n=== Complete Ranking of Reranker Models ===")
print(
    best_summary_sorted[
        [
            "reranker_model",
            "Success Rate (%)",
            "Average Context Size",
            "Average MRR",
            "Average Rerank MRR",
            "Average Mean Rank",
        ]
    ].to_string(index=False)
)
```
